# TritMorphLLM Quick Test

Load the latest experiment report and visualize perplexity plus morphology generalization.

```python
from pathlib import Path
import matplotlib.pyplot as plt
import pandas as pd
import re
import sys
root = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
sys.path.insert(0, str(root / 'src'))
print(root)
```

```python
results_path = root / 'results' / 'experiment_results.md'
markdown = results_path.read_text(encoding='utf-8')
print(markdown)
```

```python
table_lines = [line for line in markdown.splitlines() if line.startswith('|') and '---' not in line]
headers = [item.strip() for item in table_lines[0].strip('|').split('|')]
rows = [[item.strip() for item in line.strip('|').split('|')] for line in table_lines[1:]]
df = pd.DataFrame(rows, columns=headers)
df
```

```python
plot_df = df.copy()
plot_df['Held-out Perplexity'] = plot_df['Held-out Perplexity'].astype(float)
plot_df['Morph Acc'] = plot_df['Morph Acc'].astype(float)
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(plot_df['Model'], plot_df['Held-out Perplexity'], marker='o', linewidth=2)
axes[0].set_title('Held-out Perplexity')
axes[0].set_ylabel('Perplexity')
axes[0].grid(alpha=0.3)
axes[1].bar(plot_df['Model'], plot_df['Morph Acc'], color=['#2a9d8f', '#e76f51'])
axes[1].set_title('Morphology Accuracy')
axes[1].set_ylabel('Accuracy')
axes[1].set_ylim(0.0, 1.05)
for tick in axes[0].get_xticklabels() + axes[1].get_xticklabels():
    tick.set_rotation(10)
plt.tight_layout()
plt.show()
```
